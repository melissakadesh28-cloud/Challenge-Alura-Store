```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
tienda = pd.read_csv(url)

print(tienda.columns)
```

1. Análisis de facturación

```python
print("Ingreso total Tienda 1:", tienda['Precio'].sum())
print("Ingreso total Tienda 2:", tienda2['Precio'].sum())
print("Ingreso total Tienda 3:", tienda3['Precio'].sum())
print("Ingreso total Tienda 4:", tienda4['Precio'].sum())
```

2. Ventas por categoría

Más vendidas

```python
print("Categorías más vendidas Tienda 1:")
print(tienda['Categoría del Producto'].value_counts().head(5))
```

Menos vendidas

```python
print("Categorías menos vendidas Tienda 1:")
print(tienda['Categoría del Producto'].value_counts().tail(5))
```

3. Calificación promedio de la tienda

```python
print("Calificación promedio Tienda 1:", tienda['Calificación'].mean())
print("Calificación promedio Tienda 2:", tienda2['Calificación'].mean())
print("Calificación promedio Tienda 3:", tienda3['Calificación'].mean())
print("Calificación promedio Tienda 4:", tienda4['Calificación'].mean())
```

4. Productos más y menos vendidos

```python
print("Productos más vendidos Tienda 1:")
print(tienda['Producto'].value_counts().head(5))
```

```python
print("Productos menos vendidos Tienda 1:")
print(tienda['Producto'].value_counts().tail(5))
```

5. Envío promedio por tienda

```python
print("Costo promedio de envío Tienda 1:", tienda['Costo de envío'].mean())
print("Costo promedio de envío Tienda 2:", tienda2['Costo de envío'].mean())
print("Costo promedio de envío Tienda 3:", tienda3['Costo de envío'].mean())
print("Costo promedio de envío Tienda 4:", tienda4['Costo de envío'].mean())
```

```python
import matplotlib.pyplot as plt
```

 Gráfica 1: Ingreso total por tienda

```python
ingresos = [
    tienda['Precio'].sum(),
    tienda2['Precio'].sum(),
    tienda3['Precio'].sum(),
    tienda4['Precio'].sum()
]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure(figsize=(8,6))
plt.bar(tiendas, ingresos, color=['#4CAF50','#2196F3','#FFC107','#F44336'])
plt.title("Ingreso total por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingreso total")
plt.show()
```

Gráfica 2: Categorías más vendidas en Tienda 1

```python
categorias = tienda['Categoría del Producto'].value_counts().head(5)
plt.figure(figsize=(6,6))
plt.pie(categorias, labels=categorias.index, autopct='%1.1f%%', startangle=140)
plt.title("Categorías más vendidas - Tienda 1")
plt.axis('equal')
plt.show()
```

Gráfica 4: Productos más vendidos en Tienda 1

```python
productos = tienda['Producto'].value_counts().head(5)
plt.figure(figsize=(8,6))
plt.barh(productos.index, productos.values, color='skyblue')
plt.title("Productos más vendidos - Tienda 1")
plt.xlabel("Cantidad vendida")
plt.ylabel("Producto")
plt.show()
```

# **Introducción**
El presente informe tiene como propósito evaluar las cuatro tiendas del Sr. Juan considerando cinco factores clave: ingresos totales, categorías de productos más y menos vendidas, calificaciones promedio de los clientes, productos más y menos vendidos, y coste promedio de envío. El análisis busca determinar cuál tienda presenta mayores debilidades y debería ser vendida, así como identificar la tienda más rentable para establecer una comparación clara entre ambas.

# **Desarrollo**

Ingresos totales
El análisis de la columna Precio mostró que la Tienda 1 es la más rentable, con el mayor nivel de facturación. En contraste, la Tienda 4 presenta el ingreso total más bajo, reflejando baja rentabilidad y menor capacidad de sostenerse en el tiempo.

Categorías de productos
La Tienda 1 cuenta con una amplia diversidad de categorías bien posicionadas y con buena rotación. En cambio, la Tienda 4 concentra sus ventas en pocas categorías y mantiene varias con escasa participación, lo que limita su atractivo comercial.

Calificaciones promedio
La Tienda 1 mantiene una valoración alta, cercana a 4–5 estrellas, reflejando satisfacción y fidelización de clientes. Por el contrario, la Tienda 4 presenta calificaciones más bajas, lo que indica problemas en la experiencia de compra y menor confianza de los consumidores.

Productos más y menos vendidos
La Tienda 1 cuenta con varios productos líderes en ventas, con rotación constante y fuerte posicionamiento. En contraste, la Tienda 4 tiene pocos productos destacados y una larga lista de artículos con ventas mínimas, reflejando baja competitividad.

Coste promedio de envío
La Tienda 1 mantiene costes de envío competitivos, lo que favorece la percepción de los clientes y mejora la experiencia de compra. En cambio, la Tienda 4 presenta un coste promedio de envío más alto, lo que encarece la operación y reduce su atractivo frente a la competencia





# **Conclusión**
La comparación entre la Tienda 1 (más rentable) y la Tienda 4 (menos rentable) evidencia diferencias significativas en todos los factores analizados. La Tienda 1 se distingue por sus ingresos elevados, diversidad de categorías, productos líderes, alta satisfacción de clientes y costes de envío competitivos. En contraste, la Tienda 4 muestra ingresos bajos, poca diversidad, calificaciones deficientes, escasa rotación de productos y costes de envío elevados.  

Por estas razones, se recomienda al Sr. Juan vender la Tienda 4, ya que concentra las principales debilidades del negocio. Al hacerlo, podrá enfocar sus recursos en la Tienda 1, que no solo asegura mayores ingresos, sino también sostenibilidad a largo plazo gracias a su portafolio diverso y clientes satisfechos.
